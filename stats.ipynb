```python
# 24 
import numpy as np

# Data
accountancy = np.array([45, 70, 65, 30, 90, 40, 50, 75, 85, 60])
statistics = np.array([35, 90, 70, 40, 95, 40, 60, 80, 80, 50])

# Calculate means
mean_x = np.mean(accountancy)
mean_y = np.mean(statistics)

# Calculate the numerator for Pearson's correlation coefficient
numerator = np.sum((accountancy - mean_x) * (statistics - mean_y))

# Calculate the denominators
denominator_x = np.sqrt(np.sum((accountancy - mean_x) ** 2))
denominator_y = np.sqrt(np.sum((statistics - mean_y) ** 2))

# Calculate Pearson's correlation coefficient
r = numerator / (denominator_x * denominator_y)

print(f'Pearson\'s correlation coefficient: {r:.4f}')
```

```python
# 31. The mean of a distribution is 60 with a standard deviation of 10. Assuming that the dist what percentage of items be (i) between 60 and 72, (ii) between 50 and 60, (iii) beyond 72 70 and 80?
import scipy.stats as stats

# Parameters
mean = 60
std_dev = 10

# (i) Percentage between 60 and 72
lower_bound_1 = 60
upper_bound_1 = 72
percentage_between_60_and_72 = stats.norm.cdf(upper_bound_1, mean, std_dev) - stats.norm.cdf(lower_bound_1, mean, std_dev)

# (ii) Percentage between 50 and 60
lower_bound_2 = 50
upper_bound_2 = 60
percentage_between_50_and_60 = stats.norm.cdf(upper_bound_2, mean, std_dev) - stats.norm.cdf(lower_bound_2, mean, std_dev)

# (iii) Percentage beyond 72
percentage_beyond_72 = 1 - stats.norm.cdf(72, mean, std_dev)

# Print results
print(f'Percentage between 60 and 72: {percentage_between_60_and_72 * 100:.2f}%')
print(f'Percentage between 50 and 60: {percentage_between_50_and_60 * 100:.2f}%')
print(f'Percentage beyond 72: {percentage_beyond_72 * 100:.2f}%')
```

```python
# 32. 15000 students sat for an examination. The mean marks was 49 and the distribution o standard deviation of 6. Assuming that the marks were normally distributed what propor scored (a) more than 55 marks, (b) more than 70 marks
# Parameters
mean = 49
std_dev = 6

# (a) Proportion scoring more than 55
marks_a = 55
proportion_more_than_55 = 1 - stats.norm.cdf(marks_a, mean, std_dev)

# (b) Proportion scoring more than 70
marks_b = 70
proportion_more_than_70 = 1 - stats.norm.cdf(marks_b, mean, std_dev)

# Total students
total_students = 15000

# Number of students
students_more_than_55 = proportion_more_than_55 * total_students
students_more_than_70 = proportion_more_than_70 * total_students

# Print results
print(f'Proportion scoring more than 55 marks: {proportion_more_than_55:.4f}')
print(f'Number of students scoring more than 55 marks: {students_more_than_55:.0f}')
print(f'Proportion scoring more than 70 marks: {proportion_more_than_70:.4f}')
print(f'Number of students scoring more than 70 marks: {students_more_than_70:.0f}')
```

```python
# 33. If the height of 500 students are normally distributed with mean 65 inch and standard How many students have height: a) greater than 70 inch. b) between 60 and 70 inch.
# Parameters
mean_height = 65
std_dev_height = 5
total_students = 500

# (a) Number of students with height greater than 70 inches
height_a = 70
proportion_greater_than_70 = 1 - stats.norm.cdf(height_a, mean_height, std_dev_height)
students_greater_than_70 = proportion_greater_than_70 * total_students

# (b) Number of students with height between 60 and 70 inches
height_b1 = 60
height_b2 = 70
proportion_between_60_and_70 = stats.norm.cdf(height_b2, mean_height, std_dev_height) - stats.norm.cdf(height_b1, mean_height, std_dev_height)
students_between_60_and_70 = proportion_between_60_and_70 * total_students

# Print results
print(f'Number of students with height greater than 70 inches: {students_greater_than_70:.0f}')
print(f'Number of students with height between 60 and 70 inches: {students_between_60_and_70:.0f}')
```

```python
# 35.A random sample of size 25 from a population gives the sample standard derivation to hypothesis that the population standard derivation is 10.5. Hint(Use chi-square distribution).
from scipy.stats import chi2

# Parameters
sample_size = 25
sample_std_dev = 10.5
population_std_dev = 10.5
sample_variance = sample_std_dev ** 2

# Calculate the chi-square statistic
chi_square_statistic = (sample_size - 1) * (sample_variance / (population_std_dev ** 2))

# Degrees of freedom
df = sample_size - 1

# P-value
p_value = 1 - chi2.cdf(chi_square_statistic, df)

print(f'Chi-square statistic: {chi_square_statistic:.2f}')
print(f'P-value: {p_value:.4f}')
```
